```python
import numpy as np
import pandas as pd
import spacy
```

```python
nlp = spacy.load("ru_core_news_lg")
```

```python
train = pd.read_csv('./av/train.csv')
train.head()
```

```python
val = pd.read_csv('av/val.csv')
val.head()
```

```python
docs = train['description']
```

```python
shit = [[tok.lemma_ for tok in doc] for doc in nlp.pipe(docs, batch_size=512, n_process=30, disable=["parser", "ner"])]
```

```python
train['lemmatized_desc'] = [' '.join(sent) for sent in shit]
```

```python
val['lemmatied_desc'] = [' '.join([tok.lemma_ for tok in doc])
                         for doc in 
                         nlp.pipe(val['description'], batch_size=512, n_process=30, disable=["parser", "ner"])]
```

```python
val.to_csv('./av/lemmatized_val.csv', index=False)
```

```python
train.to_csv('./av/lemmatized_train.csv', index=False)
```
